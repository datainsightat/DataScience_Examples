```r
set.seed(1)
n <- 100
Sigma <- 9*matrix(c(1.0, 0.5, 0.5, 1.0), 2, 2)
dat <- MASS::mvrnorm(n = 100, c(69, 69), Sigma) %>%
  data.frame() %>% setNames(c("x", "y"))
```

```r
head(dat)
```

```r
#Q1

test_index <- createDataPartition(dat$y,times=1,p=0.5,list=FALSE)

train_set <- dat %>% slice(-test_index)
test_set <- dat %>% slice(test_index)

fit <- lm(dat$y ~ dat$x, data = train_set)
fit$coef

y_hat <- fit$coef[1] + fit$coef[2]*dat$x

avg <- mean(dat$y) #no: 69.226185562849
avg

sd <- mean((avg - dat$y)^2) #no: 7.4283897655225
sd

sqr_loss <- mean((y_hat - dat$y)^2)
sqr_loss
```

```r
#Q1

list_ <- c(1)

set.seed(1)

rmse <- function(dat,list_) {
  test_index <- createDataPartition(dat$y,times=1,p=0.5,list=FALSE)

  train_set <- dat %>% slice(-test_index)
  test_set <- dat %>% slice(test_index)
    
  fit <- lm(dat$y ~ dat$x, data = train_set)
  fit$coef

  sqr_loss <- sqrt(mean((y_hat - dat$y)^2))
  list_ <- list_ + sqr_loss
}

replicate(rmse(daet,list_),100)

head(list_)
mean(list_)
sd(list_)
```

```r
library(funModeling) 
library(tidyverse) 
library(Hmisc)
library(dslabs)
library(caret)
```

```r
data('mnist_27')
```

```r
fit <- glm(y ~ x_1 + x_2, data=mnist_27$train, family='binomial')
```

```r
p_hat <- predict(fit, newdata = mnist_27$test)
y_hat <- factor(ifelse(p_hat > 0.5, 7, 2))
confusionMatrix(data = y_hat, reference = mnist_27$test$y)
```
